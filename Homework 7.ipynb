```python
import altair as alt
import pandas as pd
import numpy as np
import json
```

```python
data = pd.read_json('C:/Users/bayardcarlson/Dropbox/Berkeley MIDS/W209 Data Visualization and Communication/Homework 6/tweets.json')
data.head()
```

```python
list_dates = list(data['created_at'])
week_days = [date.dayofweek for date in list_dates]

data['week_day'] = week_days
```

```python

# Need to get tweet counts by week_day and hour.
group_by = data.groupby(['week_day','hour']).count()
group_by
```

```python
x = np.array(group_by.index.to_frame().loc[:,'week_day'])
y = np.array(group_by.index.to_frame().loc[:,'hour'])
color = group_by['source']

final_dataset = pd.DataFrame(np.stack((x, y, color), axis = 1))
final_dataset.head()
```

```python
alt.Chart(final_dataset).mark_rect().encode(
    x= alt.X('0:N', axis=alt.Axis( title='Week Day')),
    y= alt.Y('1:N', axis=alt.Axis( title='Hour of Day')),
    color = alt.Color('2:Q', scale=alt.Scale(scheme='greenblue'), axis = alt.Axis(title = 'Tweet Count'))
).properties(title = 'Tweets by Weekday and Hour')
```
