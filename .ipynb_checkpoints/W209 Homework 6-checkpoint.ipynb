```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv("C:/Users/bayardcarlson/Dropbox/Berkeley MIDS/W209 Data Visualization and Communication/Homework 1/players_20.csv")
df.head()
```

```python
# Need to create a stacked bar chart. You could do a stacked bar chart of player counts by country with color being position. 

without_sub = df[df['team_position'] != 'SUB']
chart_data = without_sub[['nationality', 'team_position','short_name']].groupby(['nationality','team_position']).count()

nationality = np.array(chart_data.index.to_frame().loc[:,'nationality'])
position = np.array(chart_data.index.to_frame().loc[:,'team_position'])
player_count = np.array(chart_data['short_name'])

final_dataset = pd.DataFrame(np.stack((nationality, position, player_count), axis = 1))
```

```python
final_dataset
```

```python
chart = alt.Chart(final_dataset).mark_bar().encode(
    y = alt.Y('0:N', axis=alt.Axis( title='Nationality'), sort = '-x'), 
    x = alt.X('2:Q', axis=alt.Axis( title='Player Count')),
    color = alt.Color('1:N', legend=alt.Legend(title="Player Position"))
).properties(title = 'Players by Nationality and Position')

```

```python
chart_two = alt.Chart(final_dataset).mark_bar().encode(
    y = alt.Y('0:N', axis=alt.Axis( title='Nationality'), sort = '-x'), 
    x = alt.X('2:Q', axis=alt.Axis( title='Player Count')),
    color = alt.Color('1:N', legend=alt.Legend(title="Player Position")
                     , scale = alt.Scale(scheme = 'category20'))
).properties(title = 'Players by Nationality and Position')
```
